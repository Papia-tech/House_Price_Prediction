```python
# Show null value counts
df.isnull().sum()
```

```python
# Get the null value percentage
null_percentage = df.isnull().mean()*100
null_percentage
```

```python
miss_val_50_percent = null_percentage[null_percentage > 50]
miss_val_50_percent
# As per domain knowledge we will not drop those features because they have NA because of no Allay in that house, instead will will replace with constant value 'NA'
```

```python
miss_val_20_to_50_percent = null_percentage[(null_percentage > 20) & (null_percentage <51)]
miss_val_20_to_50_percent
# As per domain knowledge we will not drop those features because they have NA because of no FireplaceQu in that house, instead will will replace with constant value 'NA'
```

```python
miss_val_5_to_20_percent = null_percentage[(null_percentage > 5) & (null_percentage <21)]
miss_val_5_to_20_percent
# As per domain knowledge we will not drop those features because they have NA because of no Garage in that house, instead will will replace with constant value 'NA'
```

```python
miss_val_feature = null_percentage[null_percentage > 0]
miss_val_feature
```

```python
df["LotFrontage"].value_counts().head()
# We will not drop this feature because this is important feature
```

```python
# Catagorical null features
cat_null_feature = miss_val_feature[miss_val_feature.keys().isin(cat_features)]
cat_null_feature
```
